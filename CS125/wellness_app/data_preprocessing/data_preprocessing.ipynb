```python
import pandas as pd
import numpy as np

df = pd.read_csv('exercise_dataset.csv')
tmp0 = df['130 lb'] / 130
tmp1 = df['155 lb'] / 155
tmp2 = df['180 lb'] / 180
tmp3 = df['205 lb'] / 205
final_rates = (tmp0 + tmp1 + tmp2 + tmp3) / 4

df = df.drop(['130 lb', '155 lb', '180 lb', '205 lb', 'Calories per kg'], axis=1) 
df = pd.concat([df, final_rates], axis=1, join='inner')

df
```

```python
compression_opts = dict(method='zip',
                        archive_name='exercises_with_calories_burned_rate.csv')  
df.to_csv('out.zip', index=False,
          compression=compression_opts)
```

```python
df = pd.read_csv('exercises_post_manual_input.csv')

df
```

```python
col_max = df['Calories Burned per lb (in 60 Min)'].max()
col_min = df['Calories Burned per lb (in 60 Min)'].min()

normalized_rates = (df['Calories Burned per lb (in 60 Min)'] - col_min) / (col_max - col_min)
normalized_rates
```

```python
for index, value in enumerate(normalized_rates):
    if value > 0.66:
        df.loc[index, ('Intensity')] = 2
    elif value > 0.33:
        df.loc[index, ('Intensity')] = 1
    else:
        df.loc[index, ('Intensity')] = 0

df
```

```python
compression_opts = dict(method='zip',
                        archive_name='exercises_final.csv')  
df.to_csv('out.zip', index=False,
          compression=compression_opts)
```
